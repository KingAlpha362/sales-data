```python
from matplotlib import pyplot as plt
import pandas as pd

# a. Load the file and display the first 5 rows
df = pd.read_csv("sales-dataset.csv", encoding="latin-1")  # adjust encoding if needed

print("The shape of the data:", df.shape)
print("\nFirst 5 rows:")
print(df.head())

# b. Rename columns
df = df.rename(columns={
    "QUANTITYORDERED": "Quantity Ordered",
    "PRICEEACH": "Price Each",
    "SALES": "Sales Amount"
})

print("\nColumns after renaming:")
print(df.columns)

# c. Add a Performance column
def categorize_performance(sales):
    if sales < 1000:
        return "Low"
    elif 1000 <= sales <= 4999:
        return "Medium"
    else:
        return "High"

df["Performance"] = df["Sales Amount"].apply(categorize_performance)

print("\nFirst 5 rows with Performance column:")
print(df.head())

# d. Scatter plot of Quantity Ordered vs Sales Amount
x = df["Quantity Ordered"]
y = df["Sales Amount"]

plt.figure(figsize=(8,6))
plt.scatter(x, y, c='blue', alpha=0.5)
plt.title("Quantity Ordered vs Sales Amount")
plt.xlabel("Quantity Ordered")
plt.ylabel("Sales Amount")
plt.grid(True)
plt.show()

# Optional: describe the relationship
print("\nObservation: From the scatter plot, you can observe that as the Quantity Ordered increases, the Sales Amount generally increases, showing a positive correlation. Some variation may exist due to differences in Price Each.")

# e. Save the final dataset to a

```
